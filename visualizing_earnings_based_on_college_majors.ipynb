# Pandas, Histograms

In this portion we will explore the distribution of data points using binning to see the frequency of occurence in each column data.

```python
# One cell run 4 Histograms.
#alternative code recent_grads['Sample_size'].hist(bins=25, range=(0,5000))

columns = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize = (5, 12))

for i in range(0,4):
    ax = fig.add_subplot(4,1,1+i)
    ax = recent_grads[columns[i]].plot(kind = 'hist', rot = 35)
```

```python
# One cell run 4 Histograms.

columns = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize = (5,12))

for i in range(4,8):
    ax = fig.add_subplot(4,1,i-3) # 4 - 3 = 1, 5 - 3 = 2 . . . 
    ax = recent_grads[columns[i]].plot(kind = 'hist', rot = 35)
```

`Do students in more popular majors make more money?`
- In the first 10 majors based on popularity we can observe that the median salary range is playing in the 80 to 90 thousand USD range while the bottom ten is around 20-25 thousand USD. Therefore, There is a high possibility that students in more popular majors earn more.

`Do students that majored in subjects that were majority female make more money`
- Based on analysis this is not the case. There is no strong trend on higher median salaries and majors with majority women population. Based on observation the bottom 10 majors are populated by women. In the top 10 only astronomy and astrophysics is women populated (>50%) which earn higher median salaries.

`Is there any link between the number of full-time employees and median salary?`
- Being a fulltime employee does not always guarantee higher median salaries since a majority of the datapoints lie on 0 upto a few thousands which means that part-time employees or different arrangements can match or may even get higher salaries than fulltime employees.
